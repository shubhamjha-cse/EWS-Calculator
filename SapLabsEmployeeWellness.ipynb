### Employee Welness Score

```python
dict1={}
dict2={}
dict3={}
dict4={}
dict5={}
dict6={}
#print(data1['EmpID'][0])
for i in range(len(data1)):
    #print(i)
    dict1[data1['EmpID'][i]]=data1['Physical Fitness Checkup Score'][i]
#print(dict1)


for i in range(len(data2)):
    #print(i)
    dict2[data2['EmpID'][i]]=data2['Mental Check up Score'][i]
#print(dict2)

for i in range(len(data3)):
    #print(i)
    dict3[data3['EmpID'][i]]=data3['Corporate Wellness Metrics'][i]
#print(dict3)

for i in range(len(data4)):
    #print(i)
    dict4[data4['EmpID'][i]]=data4['Regional Quality of Life Score'][i]
#print(dict4)

for i in range(len(data5)):
    #print(i)
    dict5[data5['EmpID'][i]]=data5['Finance Metrics'][i]
#print(dict5)

for i in range(len(data6)):
    #print(i)
    dict6[data6['EmpID'][i]]=data6['WLB Score'][i]
#print(dict6)



m=3.3
p=3.3
s=2.4
d=0.0012
w=1.2
f=0.0006


list1=[]
list2=[]
for i in dict1:
    ans=p*dict1[i]*10 + m*dict2[i]*10 + s*dict3[i]*10 + d*dict4[i]*1000 + f*dict5[i]*1000 + w*dict6[i]
    list1.append(ans/12)
    list2.append([i,ans/12])


df = pd.DataFrame(list2, columns = ['EmpId', 'Wellness Score'])
```

```python
def PhysicalFitnessCheckupScoreCalculator(w1,w2,w3,w4,w5,w6,w7,w8,w9):
    ans=(math.log2(w1)*1)-(abs(w2 - 21.7)*2.2)-(abs(w3-100)*0.8)+(w4*0.5)+(w5*1)+(w6*1.3)+(w7*1.7)-(w8*0.6)+(w9*0.9)
    return ans
```

```python
def MentalHealthCheckupScoreCalculator(w1,w2,w3,w4,w5,w6,w7,w8,w9):
    ans=(math.log2(w1)*1)+(w2*3.1)-(w3*2.1)-(w4*0.5)+(w5*1.2)+(w6*0.7)+(w7*0.3)+(w8*0.3)+(w9*0.9)
    return ans
```

```python
def CorporateScoreCalculator(w1,w2,w3,w4,w5,w6,w7,w8):
    ans=(w1*2.5)+(w2*2)+(w3*1)+(w4*0.5)+(w5*1.1)+(w6*0.9)+(w7*0.5)+(w8*1.5)
    return ans
```

```python
def DemographicspScoreCalculator(w1,w2,w3,w4,w5,w6,w7):
    ans=(w1*4.5)+(w2*1.6)+(w3*0.7)+(w4*0.5)+(w5*1.2)+(w6*0.9)-(w7*0.6)
    return ans
```

```python
def FinanceScoreCalculator(w1,w2,w3,w4,w5,w6,w7,w8):
    ans=(w1*0.7)+(w2*1.5)+(w3*1.1)+(w4*1.2)+(abs(12-w5)*2)+(w6*1.1)+(w7*2)+(w8*0.4)
    return ans
```

```python
def WorkLifeScoreCalculator(w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12):
    ans=(w1*0.3)-(abs(w2 - 8)*1.5)+(w3*1)-(w4*0.5)+(w5*0.8)-(w6*1.1)-(w7*1.5)+(w8*0.6)+(w9*0.8)+(w10*0.6)+(w11*0.6)+(w12*0.7)
    return ans
```

```python
def WellnessCalculator(w1,w2,w3,w4,w5,w6):
    m=3.3
    p=3.3
    s=2.4
    d=0.0012
    f=0.0006
    w=1.2
    ans=m*w1 + p*w2 + s*w3 + d*w4 + w*w5 + f*w6
    return (ans/12)
```
